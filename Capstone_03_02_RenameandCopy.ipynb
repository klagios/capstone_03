### 2.0 Add lines to the labels for each image and copy and rename images<a id='2.0_add_copy_rename'></a>

```python
df.head()
```

```python
#Copy lines of labels for each of the images of that food
#Copy images to a single directory and rename files sequentially

new_df = df.copy()

#Get all folders in googleimages
base_dir = "googleimages"
#base_dir = "test"
folder_list = os.listdir(base_dir)

dest_dir = "data/allergens"

n = 1
dict = {}

#for each folder, get files inside as a list
for i in range(1, len(folder_list)):
    #get list of all the files in that folder
    cur_folder = base_dir + "/" + folder_list[i]
    file_list = os.listdir(cur_folder)
    
    start_n = str(n).zfill(5)
    
    #store row to copy
    copy_row = new_df.loc[folder_list[i]].copy()
    
    for j in range(len(file_list)):
        #for each image, copy it to "data/allergens" and rename it sequentially
        file = cur_folder + "/" + file_list[j]
        dest_file = dest_dir + "/" + str(n).zfill(5) + ".jpg"
        shutil.copy(file, dest_file)
        
        #propagate rows in df
        #add a line to the dataframe corresponding to that food
        if j==0:
            #if it's the first one, add the image number to the id column, and store the info from the index to the end of the non-'id' columns
            new_df.loc[folder_list[i], 'id'] = str(n).zfill(5) + '.jpg'

        else:
            #update id value
            copy_row['id'] = str(n).zfill(5) + '.jpg'
            df_row = pd.DataFrame(copy_row)

            #add to df
            #print(new_df.head())
            new_df = pd.concat([new_df, df_row.T], axis=0) # Appending new rows using concat()

        n = n+1
     

    end_n = str(n-1).zfill(5)
          
    #update dictionary with the name of the food and the numeric start and end values for the images. 
    dict[folder_list[i]] = (start_n, end_n)

#sort values by index
new_df.sort_index(inplace=True)

print(dict)
print(new_df)
```

```python
df_dict = pd.DataFrame.from_dict(dict)
df_dict
```

2.1 Write out files

```python
#Write out the new_df to csv
new_df.to_csv('data/foodallergenlabels.csv')
df_dict.to_csv('data/startendimageids.csv')
```
